## 导入外部数据
利用Pandas里的read_x()方法，x表示导入文件的格式

### 导入.xlsx文件
read_excel()

#### 基本导入
sheet_name参数

```python
# 不指定sheet_name，默认导入第一个sheet
pd.read_excel('./data/read_excel.xlsx')

# 指定sheet_name名称
pd.read_excel('./data/read_excel.xlsx', sheet_name='Sheet1')

# 指定sheet顺序，从0开始
pd.read_excel('./data/read_excel.xlsx', sheet_name=0)
```

#### 指定行索引
index_col参数，默认索引为0

```python
pd.read_excel('./data/read_excel.xlsx', index_col=0)
```

#### 指定列索引
header参数，默认值为0，即默认第一行作为列索引

```python
pd.read_excel('./data/read_excel.xlsx', header=0)
pd.read_excel('./data/read_excel.xlsx', header=1)
pd.read_excel('./data/read_excel.xlsx', header=None)
```

####  指定导入列
usecols参数指定要导入的列，传递列表导入多列

```python
pd.read_excel('./data/read_excel.xlsx', usecols=0)
pd.read_excel('./data/read_excel.xlsx', usecols=[0,2])
```

### 导入.csv文件
read_csv()

```python
# 注意编码：encoding=gb18030或utf8，默认为utf8
pd.read_csv('./data/read_csv.csv')
```

#### 指明分隔符号
默认以逗号分隔，sep参数，如：逗号、空格、制表符\t
#### 指明读取行数
nrows参数，表示前n行

```python
pd.read_csv('./data/read_csv.csv', nrows=2)
```

#### engine指定
当路径或文件包含中文（不建议）时，导入会报错，需要指定engine='Python'
#### 其他
涉及行、列索引设置及指定导入某列或某几列，方法与导入.xlsx文件一致

### 导入.txt文件
read_table()方法，也可以导入.csv文件，需要指定seq参数，其他用法和read_csv()相同

```python
df = pd.read_table('./data/read_txt.txt', sep=' ')
```

### 导入sql文件
pymysql模块，read_sql()方法

```python
# user: 用户名
# password：密码
# host：数据库地址/本机使用localhost
# db：数据库名
# charset：数据库编码，一般为utf8
conn = pymysql.connect(host='localhost', user='user', password='password', db='db', charset='utf8')
sql = 'select * from law order by crt_time desc limit 0,2'
pd.read_sql(sql, conn)
```
