## 导入外部数据

### 导入.xlsx文件

#### 基本导入
sheet_name参数

```python
# 不指定sheet_name，默认导入第一个sheet
pd.read_excel('./data/read_excel.xlsx')

# 指定sheet_name名称
pd.read_excel('./data/read_excel.xlsx', sheet_name='Sheet1')

# 指定sheet顺序，从0开始
pd.read_excel('./data/read_excel.xlsx', sheet_name=0)
```

#### 指定行索引
index_col参数，默认索引为0

```python
pd.read_excel('./data/read_excel.xlsx', index_col=0)
```

#### 指定列索引
header参数，默认值为0，即默认第一行作为列索引

```python
pd.read_excel('./data/read_excel.xlsx', header=0)
pd.read_excel('./data/read_excel.xlsx', header=1)
pd.read_excel('./data/read_excel.xlsx', header=None)
```
